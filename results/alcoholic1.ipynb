```python
# parse the jason data

def isfloat(value):
  try:
    float(value)
    return True
  except:
    return False

lat=np.array([])
long=np.array([])
#lat=np.array(data[:]['Lat'])
for irow in range(21253-10):
    #print irow,data[irow]['Lat']
    # if irow>100 : break
    
    if isfloat(data[irow]['Lat']) & isfloat(data[irow]['Long']):
        lat=np.append(lat,float(data[irow]['Lat']))
        long=np.append(long,float(data[irow]['Long']))
print lat[0:10]
print long[0:10]
```

```python
print "lat1=",lat[0:10]
print "long1=",long[0:10]
plt.scatter(lat,long)
plt.show()
```
